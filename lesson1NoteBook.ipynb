```python
import numpy as np
# יצירת מערך של 10 מספרים שלמים בטווח [1, 50]
arr1 = np.random.randint(1, 51, size=10)
print(f"המערך: {arr1}")

# חישוב הממוצע
mean_val = np.mean(arr1)
print(f"הממוצע של המספרים: {mean_val}")


```

```python
# יצירת שני מערכים דו-ממדיים (מטריצות) בגודל 3x3 עם מספרים רנדומליים
matrix_a = np.random.randint(1, 11, size=(3, 3))
matrix_b = np.random.randint(1, 11, size=(3, 3))

# חיבור המערכים
matrix_sum = matrix_a + matrix_b

print("מערך א':\n", matrix_a)
print("מערך ב':\n", matrix_b)
print("תוצאת החיבור:\n", matrix_sum)

# בדיקת הממוצע של מערך התוצאה
total_mean = np.mean(matrix_sum)
print(f"\nהממוצע של תוצאת החיבור: {total_mean}")
```

```python
# יצירת מערך של 20 מספרים
arr3 = np.random.randint(1, 101, size=20)
print(f"המערך המלא: {arr3}")

# מציאת המספרים הגדולים מ-50 באמצעות Boolean Indexing
greater_than_50 = arr3[arr3 > 50]
print(f"מספרים שגדולים מ-50: {greater_than_50}")
```

```python
# יצירת מערך עם מספרים מ-1 עד 10
arr4 = np.arange(1, 11)
print(f"המערך המקורי: {arr4}")

# ערבוב המערך (In-place)
np.random.shuffle(arr4)
print(f"המערך לאחר ערבוב: {arr4}")

# בחירת אלמנטים אקראיים מתוך המערך (למשל 3 אלמנטים)
random_selection = np.random.choice(arr4, size=3, replace=False)
print(f"בחירה אקראית של 3 אלמנטים: {random_selection}")
```
